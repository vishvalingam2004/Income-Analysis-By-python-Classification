```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
c=["income_above_limit"]
encoder=OrdinalEncoder()
df[c]=encoder.fit_transform(df[c])
df
```

```python
df['income_above_limit'].unique()
```

```python
x=df[["ID","age","employment_stat","wage_per_hour","working_week_per_year","industry_code","occupation_code","total_employed","vet_benefit","gains","losses","stocks_status","mig_year","importance_of_record"]]
y=df["income_above_limit"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
x_train_ID=x_train["ID"]
x_train=x_train.drop(["ID"],axis=1)
print(x_train_ID,x_train)
model.fit(x_train,y_train)
```

```python
x_test_ID=x_test["ID"]
x_test=x_test.drop(["ID"],axis=1)
y_pred=model.predict(x_test)
res=pd.DataFrame({'ID':x_test_ID,'income_above_limit':y_pred})
print(res)
```
